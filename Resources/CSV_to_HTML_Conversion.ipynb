# Converting the CSV file to an HTML Table

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cities_df = pd.read_csv('cities.csv')
cities_df
```

```python
with open('cities.html', 'w') as cities_html:
    cities_df.to_html(cities_html)
```

# Generating the Graph Images

## Temperature vs. Latitude

```python
x_axis_lat = cities_df["Lat"]
y_axis_temp = cities_df["Max Temp"]

plt.scatter(x_axis_lat, y_axis_temp)
plt.title("Maximum Temperature vs. Latitude, 6/13/2018")
plt.xlabel("Latitude (degrees)")
plt.ylabel("Maximum Temperature (degrees Fahrenheit)")
plt.grid(True)

plt.savefig("../assets/images/max_temp.png")
plt.show()
```

## Cloudiness vs. Latitude

```python
y_axis_cloud = cities_df["Cloudiness"]

plt.scatter(x_axis_lat, y_axis_cloud)
plt.title("Cloudiness vs. Latitude, 6/13/2018")
plt.xlabel("Latitude (degrees)")
plt.ylabel("Cloudiness (%)")
plt.grid(True)

plt.savefig("../assets/images/cloudiness.png")
plt.show()
```

## Humidity vs. Latitude

```python
y_axis_hum = cities_df["Humidity"]

plt.scatter(x_axis_lat, y_axis_hum)
plt.title("Humidity vs. Latitude, 6/13/2018")
plt.xlabel("Latitude (degrees)")
plt.ylabel("Relative Humidity (%)")
plt.grid(True)

plt.savefig("../assets/images/humidity.png")
plt.show()
```

## Wind Speed vs. Latitude

```python
y_axis_wind = cities_df["Wind Speed"]

plt.scatter(x_axis_lat, y_axis_wind)
plt.title("Wind Speed vs. Latitude")
plt.xlabel("Latitude (degrees)")
plt.ylabel("Wind Speed (mph)")
plt.grid(True)

plt.savefig("../assets/images/wind_speed.png")
plt.show()
```
